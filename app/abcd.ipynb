```python
# Step 1: Create list of (mobile_no, department) from each df
list1 = list(zip(df['Mobile No'], df['Departments']))
list2 = list(zip(df1['Mobile No'], df1['Departments']))
list3 = list(zip(df2['Mobile No'], df2['Departments']))

# Step 2: Combine all
combined = list1 + list2 + list3

# Step 3: Check duplicates
from collections import Counter
counter = Counter(combined)

# Step 4: Print or collect duplicates
duplicates = [item for item, count in counter.items() if count > 1]

print("Duplicates found:", duplicates)
```
